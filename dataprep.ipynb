```python
#Remove cuisines with less than 50 recipes
cuisine_counts = recipes["cuisine"].value_counts()
cuisines_to_keep_indices = cuisine_counts > 50
cuisines_to_keep = np.array(cuisine_counts.index.values)[np.array(cuisines_to_keep_indices)]
```

```python
print("Rows before {}".format(recipes.shape[0]))
recipes = recipes.loc[recipes["cuisine"].isin(cuisines_to_keep)]
print("Rows after {}".format(recipes.shape[0]))
```

```python
recipes = recipes.replace(to_replace='Yes', value=1)
recipes = recipes.replace(to_replace='No',value=0)
```

```python
#Profile of each cuisine
cuisines = recipes.groupby("cuisine").mean()
print(type(cuisines))
cuisines.head()
```

```python
#Four most common ingredients
num_ingredients = 4

# define a function that prints the top ingredients for each cuisine
def print_top_ingredients(row):        
    print(row.name.upper())
    row_sorted = row.sort_values(ascending=False)*100
    ingredients_row_sorted = list(row_sorted.index.values)[0:num_ingredients]
    row_sorted = list(row_sorted)[0:num_ingredients]
    
    for ind, ingredient in enumerate(ingredients_row_sorted):
        print("%s (%d%%)" % (ingredient, row_sorted[ind]), end=' ')
    print("\n")
    
cuisines.apply(print_top_ingredients, axis=1)
```
